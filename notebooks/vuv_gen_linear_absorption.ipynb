# Introduction

We are interested in generating 118 nm light in order to ionize and detect the OH molecule in our experiments. In order to do this, we triple a ~7 ns long pulse of 355 nm light using harmonic generation. Conversion from 355 nm to 118 nm is done in xenon, which has a large nonlinear susceptibility at 118 nm due to the transition in xenon from the 5p ground state to the 5d excited state at 119.2 nm.

One factor that may cause problems in this process is that the very transition that we rely on to convert 355 nm light to 118 nm light may also absorb significant amounts of 118 nm light. The light produced at ~118 nm is far from the transition at 119.2 nm, but broadening of the 119.2 nm transistion may be sufficient to absorb a sizable fraction of 118 nm light.

In thhis notebook, we look at how broadening of the transitions in xenon near 118 nm may cause direct absorption of 118 nm photons.

# Linear Absorption

There are two transitions in Xenon that are fairly close to 118.2 nm: the Xe $(^2 P_{3/2})5d [\frac{3}{2}]$ line at 119.203 nm, and the Xe $(^2 P_{3/2})7s$ line at 117.043 nm. Each of these lines will have a number of broadening mechanisms that may allow absorption at 118.2 nm. These methods are throught the natural linewidth, the doppler broadening from thermal motion, and pressure broadening through collisions with similar and different atoms. 

```python
lambda_5d = 119.203E-9  # m
lambda_7s = 117.043E-9  # m

omega_5d = lambda_to_freq(lambda_5d)
omega_7s = lambda_to_freq(lambda_7s)
```

## Natural Linewidth

Data for the 5d linewidth is from:
https://www.sciencedirect.com/science/article/abs/pii/0009261477805320

We can assume that the lifetime of the 7s state is on the order of a few nanoseconds. We will see shortly that error in this assumption will not affect the overall conclusions.

```python
xe_7s_lifetime = 5E-9  # [s]
xe_5d_lifetime = 1.4E-9 # [s]
```

```python
def natural_linewidth(lifetime):
    "Calculate the FWHM linewidth of a Lorentzian distribution."
    return 1 / (2 * np.pi * lifetime)

xe_7s_natural_linewidth = natural_linewidth(xe_7s_lifetime)  # Hz
xe_5d_natural_linewidth = natural_linewidth(xe_5d_lifetime) # Hz
```

## Doppler Broadening

```python
def doppler_broadening(transition_freq, temperature, mass):
    "Calculate the std. dev. of the gaussian for doppler broadening."
    return transition_freq/c * np.sqrt(2 * kB * temperature / mass)
    

xe_5d_doppler_linewidth = doppler_broadening(omega_5d, 295, mXe)  # Hz
xe_7s_doppler_linewidth = doppler_broadening(omega_7s, 295, mXe)  # Hz
```

## Collisional Broadening

Collisional broadening occurs due to the enhanced quenching of an excited species when it collides with a foreign body. This causes an effective decrease in the lifetime of the excited state, which causes an increase in the uncertainty of the energy of the transition, broadening the line.

When two different atoms collide, such as Ar and Xe, we will assume that they do not have any matching dipole allowed transitions to the ground state that would artificially increase their collisional cross section. For collisions of this type, we will use a standard pressure broadening value of 5-10 MHz/Torr.

When two atoms of the same species get close, and one of the excited states has a dipole allowed transition to the ground state, their collisional cross section is greatly enhanced by the resonance, and their effective pressure broadening is much larger. Xenon's 5d and 7s lines both are dipole allowed transitions to the ground state, and have a large, "self-pressure broadening." To find these values, we will crudely extract a fit from the data presented in https://journals.aps.org/pra/abstract/10.1103/PhysRevA.36.81. Data points from Fig. 11 (5d) and Fig. 14 (7s) were extracted with WebPlotDigitizer:

```python
collision_data_path = "../data/external/"

col_names = ['Pressure', 'Broadening']
xe_5d_collision_data = pd.read_csv(collision_data_path + "Xenon_5d_Collision.csv",
                                  header=0, names=col_names)
xe_7s_collision_data = pd.read_csv(collision_data_path + "Xenon_7s_Collision.csv",
                                  header=0, names=col_names)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(9, 4), dpi=150)

ax[0].set_title("5d Xe-Xe Collisional Broadening")
ax[1].set_title("7s Xe-Xe Collisional Broadening")

xe_5d_collision_data.plot(x='Pressure', y='Broadening', ax=ax[0], kind='scatter')
xe_7s_collision_data.plot(x='Pressure', y='Broadening', ax=ax[1], kind='scatter')

for axis in ax:
    axis.set_xlabel("Xe pressure (Torr)")
    axis.set_ylabel("Broadening (nm)")
    
plt.tight_layout()
plt.show()
```

This data is influenced by a the laser width of the measurement laser, which is reported to be 0.20 $cm^{-1}$ or 0.0025 nm. We can find the self broadening parameter by fitting:

$(\Delta \lambda_{measured})^2 = (\Delta \lambda_{laser})^2 + (\Delta \lambda_{broadening})^2$

We will also artificially limit the regime that we are fitting to pressures below 500 Torr. This is because the data from the paper begins to have larger error bars above 500 Torr, and we only anticipate working in the lower-pressure regime.

```python
P_thres = 500  # Torr

xe_5d_coll_data = xe_5d_collision_data[xe_5d_collision_data.Pressure<=P_thres]
xe_7s_coll_data = xe_7s_collision_data[xe_7s_collision_data.Pressure<=P_thres]
```

```python
delta_lambda_laser = 0.0025 # nm

def quad_fit_func(x, delta_lambda_coll):
    return np.sqrt(delta_lambda_laser**2 + (delta_lambda_coll*x)**2)

def fit_delta_lambda_coll(df):
    return curve_fit(quad_fit_func, df['Pressure'], df['Broadening'])

popt_5d, pcov_5d = fit_delta_lambda_coll(xe_5d_coll_data)
popt_7s, pcov_7s = fit_delta_lambda_coll(xe_7s_coll_data)
```

We can plot our fits to verify that they are reasonable given the data.

```python
pressures = np.linspace(0, P_thres, 1000)

fig, ax = plt.subplots(1, 2, figsize=(9, 4), dpi=150)

ax[0].set_title("5d Xe-Xe Collisional Broadening")
ax[1].set_title("7s Xe-Xe Collisional Broadening")

xe_5d_coll_data.plot(x='Pressure', y='Broadening', ax=ax[0], kind='scatter')
ax[0].plot(pressures, quad_fit_func(pressures, *popt_5d), 'r-')

xe_7s_coll_data.plot(x='Pressure', y='Broadening', ax=ax[1], kind='scatter')
ax[1].plot(pressures, quad_fit_func(pressures, *popt_7s), 'r-')

for axis in ax:
    axis.set_xlabel("Xe pressure (Torr)")
    axis.set_ylabel("Broadening (nm)")
    
plt.tight_layout()
plt.show()
```

Visual confirmation of the fits shows that the y interecept for the 5d transition data is probably too high, but that the y-intercept for the 7s data looks decent. 

```python
def quad_fit_func2(x, delta_lambda_coll, delta_lambda_laser):
    return np.sqrt(delta_lambda_laser**2 + (delta_lambda_coll*x)**2)

def fit_delta_lambda_coll2(df):
    return curve_fit(quad_fit_func2, df['Pressure'], df['Broadening'])

popt_5d_2param, pcov_5d_2param = fit_delta_lambda_coll2(xe_5d_coll_data)
popt_7s_2param, pcov_7s_2param = fit_delta_lambda_coll2(xe_7s_coll_data)
```

```python
pressures = np.linspace(0, P_thres, 1000)

fig, ax = plt.subplots(1, 2, figsize=(9, 4), dpi=150)

ax[0].set_title("5d Xe-Xe Collisional Broadening")
ax[1].set_title("7s Xe-Xe Collisional Broadening")

xe_5d_coll_data.plot(x='Pressure', y='Broadening', ax=ax[0], kind='scatter')
ax[0].plot(pressures, quad_fit_func2(pressures, *popt_5d_2param), 'r-')

xe_7s_coll_data.plot(x='Pressure', y='Broadening', ax=ax[1], kind='scatter')
ax[1].plot(pressures, quad_fit_func2(pressures, *popt_7s_2param), 'r-')

for axis in ax:
    axis.set_xlabel("Xe pressure (Torr)")
    axis.set_ylabel("Broadening (nm)")
    
plt.tight_layout()
plt.show()
```

```python
print("5d Transition:")
print(f"Pressure broadening one parameter: {popt_5d[0]:.1E} nm/Torr.")
print(f"Pressure broadening two parameter: {popt_5d_2param[0]:.1E} nm/Torr.")
print(f"Laser linewidth one parameter {delta_lambda_laser:.1E} nm (given in paper).")
print(f"Laser linewdith two parameter {popt_5d_2param[1]:.1E} nm. \n")

print("7s Transition:")
print(f"Pressure broadening one parameter: {popt_7s[0]:.1E} nm/Torr.")
print(f"Pressure broadening two parameter: {popt_7s_2param[0]:.1E} nm/Torr.")
print(f"Laser linewidth one parameter {delta_lambda_laser:.1E} nm (given in paper).")
print(f"Laser linewdith two parameter {popt_7s_2param[1]:.1E} nm.")
```

Visually, the two parameter fit is much better than the one parameter fit. While the 7s data agrees with the given laser linewidth of 0.0025 nm, the 5d data seems to have been taken when the laser was in a happy place, and with a narrower linewidth. Alternatively, by truncating the original data set at 500 Torr, we aren't fitting all of the data, and that is introducing error into the fits.

For now, I am going to move forward with the 2 parameter fits because they fit much better. We can look at what difference using the single parameter fits makes later if it becomes necessary.

We now need to convert the fitted parameter, which is in $\Delta nm$ from the center of the laser line to frequency so that we can compare to our other broadening mechanisms.

```python
def delta_lamb_to_delta_omega(lamb, delta_lamb):
    """
    Assumes delta_lamb << lamb
    """
    return c * delta_lamb / (lamb**2)

xe_5d_collision_linewidth = delta_lamb_to_delta_omega(lambda_5d, popt_5d_2param[0]*1E-9)
xe_7s_collision_linewidth = delta_lamb_to_delta_omega(lambda_7s, popt_7s_2param[0]*1E-9)
```

## Putting it all together

Steck, Eq. 1.30, p.32

```python
def gaussian(x, x0, sigma):
    return 1/np.sqrt(2*np.pi*sigma**2) * np.exp(-(x-x0)**2/(2*sigma**2))  

def lorentzian(x, x0, gamma):
    """
    gamma: HWHM
    """
    return 1/(np.pi * gamma) * gamma**2 / ((x-x0)**2 + gamma**2)
    
def _voigt_f(sigma, gamma):
    return (sigma**5 
     + 2.69269*sigma**4 * gamma
     + 2.42843 * sigma**3 * gamma**2 
     + 4.47163 * sigma**2 * gamma**3 
     + 0.07842 * sigma * gamma**4 
     + gamma**5)**(1/5)
    
def _voigt_eta(sigma, gamma):
    return (1.36603 * (gamma / _voigt_f(sigma, gamma)) 
    - 0.47719 * (gamma / _voigt_f(sigma, gamma))**2 
    + 0.11116 * (gamma / _voigt_f(sigma, gamma))**3)
    
def voigt_aprx(x, x0, sigma, gamma):
    return (_voigt_eta(sigma, gamma) * lorentzian(x, x0, gamma)
           + (1 - _voigt_eta(sigma, gamma)) * gaussian(x, x0, sigma))
```

```python
def cross_section(frequency, pressure):
    prefactor = 2 * e_charge**2 / (m_electron * eps_0 * c)
    f_5d = 0.379  # https://journals.aps.org/pra/pdf/10.1103/PhysRevA.46.149
    f_7s = 0.0859
    
    gamma_5d = xe_5d_natural_linewidth + xe_5d_collision_linewidth * pressure
    sigma_5d = xe_5d_doppler_linewidth
    x0_5d = omega_5d + xe_5d_shift_linewidth * pressure
    voigt_5d = voigt_aprx(frequency, x0_5d, sigma_5d, gamma_5d)
    
    gamma_7s = xe_7s_natural_linewidth + xe_7s_collision_linewidth * pressure
    sigma_7s = xe_7s_doppler_linewidth
    voigt_7s = voigt_aprx(frequency, omega_7s, sigma_7s, gamma_7s)
    
    return prefactor * (f_5d * voigt_5d + f_7s * voigt_7s)
```

```python
def cross_section_plot():
    
    pressures = [(1, 'tab:purple'),
                 (5, 'tab:blue'), 
                 (25, 'tab:green'),
                 (100, 'tab:orange')]
    lamb_range = np.linspace(116.5E-9, 
                             119.5E-9, 
                             10000)
    freq_range = lambda_to_freq(lamb_range)

    fig_sig, ax = plt.subplots(1, 1, figsize=(9, 6), dpi=150)

    for p, color in pressures:
        ax.semilogy(lamb_range*1E9, cross_section(freq_range, p)*10**4, color=color,
                   label = (str(p) + " Torr"))

    ax.set_xlabel("Wavelength (nm)")
    ax.set_ylabel("Atomic absorption cross-section (cm^2)")
    ax.set_ylim([1E-20, 1E-14])

    ax.axvline(119.203, linestyle='--', linewidth=1, color='k')
    ax.axvline(117.043, linestyle='--', linewidth=1, color='k')

    ax.legend()
    
    return ax

cross_section_plot()
plt.show()
```

## Error in pump frequency

We can compare $\frac{d \sigma}{d \lambda}$ and $\frac{d \sigma}{d P_{Xe}}$ and show that the difference in $\lambda$ is much less important (I think) than the variation in pressure.

```python
NdYAG_lamb_bounds = np.array([1063, 1065])/9 * 1E-9
lamb_range = np.linspace(NdYAG_lamb_bounds[0], NdYAG_lamb_bounds[1], 1000)
freq_range = np.expand_dims(lambda_to_freq(lamb_range), axis=1)
dfreq = freq_range[1, 0]-freq_range[0, 0]

pressures = np.expand_dims(np.linspace(1, 100, 500), axis=0)
dpressure = pressures[0, 1] - pressures[0, 0]

cross_section_grid = cross_section(freq_range, pressures)
dsig_df, dsig_dp = np.gradient(cross_section_grid, dfreq, dpressure)

print(f"Cross section changes by {np.mean(dsig_dp):.2E} cm^2 per Torr pressure increase.")
print(f"Cross section changes by {np.mean(dsig_df):.2E} cm^2 per Hz of frequency increase.")
freq_per_torr = np.mean(dsig_dp)/np.mean(dsig_df)
print(f"This means one Torr of pressure increase is the equivalent of changing the frequency by "
      f"{freq_per_torr*1E-9:.2F} GHz")
delta_lamb_per_torr = freq_to_lambda(omega_118 + freq_per_torr) - lambda_118
print(f"That is equivalent to ~{delta_lamb_per_torr*1E9:.4F} nm per Torr.")
```

The change in cross section based on small errors in the wavelength of the 118 nm light are on the same order as te changes inn pressure that we are using for this experiment, if not a little smaller. While narrowing down the exact wavelength of the 355 nm pump laser would help us pin this parameter down, a little error here shouldn't drastically change our conclusions.

## Lorentzian approximation in the tails

Far from the center of the absorption line we can assume it is a lorentzian. Very far from the center, we can make approximations that allow us to show that cross section is essentially linear in $\gamma$.

From Steck, if the cross-section is $a$:

$a(\omega) = \frac{N e^2}{m \epsilon_0 c \gamma} \frac{(\gamma/2)^2}{(\omega-\omega_0)^2 + (\gamma/2)^2}$

If $\gamma$ is definied as te FWHM.

If we assume that we are far in the tails of the distribution: $(\omega-\omega_0) >> \gamma$, then we can approximate:

$(\omega-\omega_0)^2+(\gamma/2)^2 = (\omega-\omega_0)^2$

and that leaves us with:

$a(\omega) = \frac{N e^2}{4 m \epsilon_0 c} \frac{\gamma}{(\omega-\omega_0)^2}$

Which means, in the tails of the distribution, the cross section is linear with the width of the distribution.
