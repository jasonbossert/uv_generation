# Parameters for a 3D Sim

```python
simulation_params = {
'b': 0.023,  # [m] Confocal parameter taken from 118 OH paper
'beam_energy': 0.017,  # [J] single pulse energy
'duration': 7*nano,   # [s] single pulse length
'dsx': 8,  # downsampling rate for x
'dsy': 8,
'dsz': 1,
'z_max': 40*milli,  # [m] length of simulation 
#'z_max': 3*milli,  # [m] length of simulation 
'Nz': 32,  # cells in z direction
'lamb355': lamb355, # [m] wavelength of 355
'lamb118': lamb118, # [m] wavelength of 118
'x_max': 300*micro, # [m] width of simulation in x and y
#'x_max': 20*micro, # [m] width of simulation in x and y
'chi2': 2.91508556e-36,  # [m/V] second-order loss constant
'chi3': 1.5e-44, # [m^2/V^2] third-order gain constant
'PXe': 25*torr, # [torr] simulation pressure
'exponent': 2
}

class Params(object):  
    def __init__(self, adict):
        self.__dict__.update(adict)
    
params = Params(simulation_params)
```

```python
def process_118(beam118):
    fluorescence = np.sum(np.abs(beam118.loss[:,:,:])**2)
    signal = np.sum(np.abs(beam118.E_xz[:,:,-1])**2)
    return (signal, fluorescence)
```

# Pressure Scan

```python
PXes = np.array([1, 5, 10, 15, 20, 25, 30, 40, 60, 80])*torr
scan_parameter(params, "PXe", PXes, "PXe_scan_sigma_fit/")
```

## Find Useful files

```python
# load csv directory
directory_name = "PXe_scan_sigma_fit/"
data_table_name = "parameter_scan_PXe"

data_lookup = pd.read_csv((directory_name + data_table_name + ".csv"))
data_lookup = data_lookup.set_index(['b', 'PXe'])
data_lookup.loc[0.023].head()
```

```python
# search CSV for meaningful parameters, return file 
bs = sorted(list(set([item[0] for item in data_lookup.index])))
#bs.remove(0.23)
P = sorted(list(set([item[1]/torr for item in data_lookup.index])))
signal_all = []
fluorescence_all = []
for b in bs:
    signal = []
    fluorescence = []
    filenames = list(data_lookup.loc[b].filename)
    for filename in filenames:
        beam355, beam118, _ = load_beams("" + filename)
        sig, fluo = process_118(beam118)
        signal.append(sig)
        fluorescence.append(fluo)
    signal_all.append(signal)
    fluorescence_all.append(fluorescence)
```
